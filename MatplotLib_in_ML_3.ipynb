```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
profit = [10, 20, 30, 25]
labels = ["Amazon", "Flipkart", "Snapdeal", "JioMart"]

colors = ["red", "magenta", "pink", "blue"]
explode = [0.05, 0.1, 0, 0]

plt.pie(profit, colors= colors, labels= labels, autopct= "%.2f%%", explode= explode, counterclock= False, startangle=90) #by default, piechart starts in counterclockwise direction and always start where the time position is 3'o clock.
plt.show()
```
