```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
x2 = [1,2,3,4,5]
y = [1,2,3,4,5]
y2 = [5,4,3,2,1]

plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
x = [1,2,3,4,5]
x2 = [1,2,3,4,5]
y = [1,2,3,4,5]
y2 = [5,4,3,2,1]

plt.plot(x,y, "r-.")
plt.plot(x2,y2)
plt.show
```

```python
import numpy as np

x = np.array([1,2,3,4,5,6])
y = 2*(x**4) + 100
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = 2*(x**4)+100

plt.plot(x,y)
plt.show()
```
